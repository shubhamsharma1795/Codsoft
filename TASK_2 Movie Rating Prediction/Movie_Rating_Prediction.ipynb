```python
movies_data = movies_data.dropna(subset =['Votes'])
movies_data.head()
```

**MODEL TRAINING AND TESTING**

```python
from sklearn.model_selection import train_test_split
movies_data['Year'] = movies_data['Year'].astype(str)
movies_data['Duration'] = movies_data['Duration'].astype(str)
movies_data['Year'] = movies_data['Year'].str.extract('(\d+)').astype(float)
movies_data['Duration'] = movies_data['Duration'].str.extract('(\d+)').astype(float)
x = movies_data[['Year','Duration','Votes']]
y = movies_data['Rating']
x_train , x_test, y_train , y_test = train_test_split(x , y , test_size = 0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train , y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
mae= mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared= False)
r2 = r2_score(y_test, y_pred)
print(f"Mean Absolute Error: {mae}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R-squared(R2) Score: {r2}")
```

**Prediction in Visualizing for Better Understanding**

```python
y_test = np.random.rand(100)
y_pred = np.random.rand(100)
errors = y_test - y_pred
fig , axs = plt.subplots(3, 1 , figsize = (8,12))

axs[0].scatter(y_test, y_pred)
axs[0].set_xlabel("Actual Rating")
axs[0].set_ylabel("Predicted Rating")
axs[0].set_title("Actual Rating vs Predicted Rating")

# Line Plot
movies_sample = np.arange(1, len(y_pred) +1)
axs[1].plot(movies_sample, y_pred, marker = 'o',linestyle = '-')
axs[1].set_xlabel("Movie Sample")
axs[1].set_ylabel("Predicted Rating")
axs[1].set_title("Predicted Rating across Movies Sample")
axs[1].tick_params(axis = 'x', rotation = 45)


# Histogram


axs[2].hist(errors , bins = 30)
axs[2].set_xlabel("Prediction Errors")
axs[2].set_ylabel("Frequency")
axs[2].set_title("Distribution of Frequency Errors")
axs[2].axvline(x = 0, color = 'r', linestyle ='--')
plt.tight_layout()
plt.show()
```
